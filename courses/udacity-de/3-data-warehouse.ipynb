````python
%%sql 
SELECT DISTINCT f.title, CONCAT(a.first_name, ' ', a.last_name)
FROM film_actor fa 
JOIN film f ON fa.film_id =f.film_id
JOIN actor a ON fa.actor_id = a.actor_id
LIMIT 10
````
